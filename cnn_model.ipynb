```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

Load CIFAR-10 dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

Print dataset shapes


```python
print("Training Images Shape :", x_train.shape)
print("Training Labels Shape :", y_train.shape)

print("Testing Images Shape :", x_test.shape)
print("Testing Labels Shape :", y_test.shape)
```

Normalize image pixel values

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

Creating CNN modela


```python
model = Sequential([

    Conv2D(32, (3,3), padding='same', activation='relu', input_shape=(32,32,3)),
    BatchNormalization(),
    Conv2D(32, (3,3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.2),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.3),
    Conv2D(128, (3,3), padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(128, (3,3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.4),

    Flatten(),

    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(10, activation='softmax')

])
```

Compile the model

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5,restore_best_weights=True)
```

Evaluate the model on test data

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```

```python
print("\nFinal Test Accuracy :", test_accuracy)
```

Model Architecture Details

The Convolutional Neural Network (CNN) model was designed from scratch for multi-class image classification using the CIFAR-10 dataset. The architecture contains convolution layers, batch normalization, max pooling layers, dropout layers, flatten layer, and fully connected dense layers.

The convolution layers are responsible for extracting image features such as edges, textures, and patterns. Batch normalization improves training stability and learning speed. Max pooling reduces feature map dimensions and computational cost. Dropout layers help reduce overfitting during training. Finally, dense layers perform classification and the Softmax output layer predicts probabilities for the 10 image classes.

   Explanation of Approach

   
In this project, a Convolutional Neural Network (CNN) was implemented from scratch using the CIFAR-10 dataset for multi-class image classification. The dataset contains 10 different image classes, and each image was resized to the shape (32,32,3). The pixel values were normalized by dividing them by 255 to improve training stability and model performance.

The labels were converted into categorical format using one-hot encoding to ensure proper tensor shape handling during training. The CNN architecture was built using multiple convolution layers for feature extraction, batch normalization layers for stable learning, max pooling layers for dimensionality reduction, and dropout layers to reduce overfitting.

The model was compiled using the Adam optimizer and categorical crossentropy loss function. During training, early stopping was used to stop the training process when validation performance stopped improving. Finally, the model was evaluated on the test dataset and achieved good classification accuracy while maintaining a simple and efficient architecture without using any pretrained models.
